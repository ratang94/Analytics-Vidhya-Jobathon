```python
import pandas as pd
```

##### Make the first column("ID") as index column

```python
df = pd.read_csv('train.csv',index_col=0)
```

##### Display the first 5 rows to check the data

```python
df.head()
```

##### Created few functions to perform data preprocessing and impute null values

```python
def signup(x):
    if(pd.isnull(x)):
        return 'No_Value'
    else:
        return (x[:4])
def camp_var1(x):
    if(x<9):
        return 10
    elif(x<13):
        return 4
    else:
        return 1
def camp_var2(x):
    if(x<7):
        return 10
    elif(x<10):
        return 7
    else:
        return 1
def buyer(x):
    if(pd.isnull(x)):
        return 0
    elif(x<3):
        return int(5)
    else:
        return int(10)
def created(x):
    return(int(x[5:7]))
```

```python
df['signup_date']=df['signup_date'].apply(signup)
```

```python
df['campaign_var_1']=df['campaign_var_1'].apply(camp_var1)
```

```python
df['campaign_var_2']=df['campaign_var_2'].apply(camp_var2)
```

```python
df['products_purchased']=df['products_purchased'].apply(buyer)
```

```python
df['created_at']=df['created_at'].apply(created)
```

##### Creating dummy variable i.e OneHotEncoding

```python
dum1=pd.get_dummies(df['campaign_var_1'],prefix='campaign_var_1')
```

```python
dum2=pd.get_dummies(df['campaign_var_2'],prefix='campaign_var_2')
```

```python
dum3=pd.get_dummies(df['signup_date'],prefix='signup')
```

```python
dum4=pd.get_dummies(df['products_purchased'],prefix='products_purchased')
```

##### Adding the extra columns to the data table

```python
df=pd.concat([df,dum1,dum2,dum3,dum4],axis=1)
```

```python
df
```

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split
```

##### I've omitted the created_at column as it was having negative effects on the model and it's not much of an influencer

```python
X = df[['user_activity_var_1','user_activity_var_2',
       'user_activity_var_3','user_activity_var_4','user_activity_var_5', 'user_activity_var_6', 'user_activity_var_7',
       'user_activity_var_8','user_activity_var_9','user_activity_var_11','user_activity_var_10','user_activity_var_12','campaign_var_1_1', 'campaign_var_1_4',
       'campaign_var_1_10', 'campaign_var_2_1', 'campaign_var_2_7',
       'campaign_var_2_10', 'signup_2015', 'signup_2016', 'signup_2017',
       'signup_2018', 'signup_2019', 'signup_2020', 'signup_2021',
       'signup_2022', 'signup_No_Value', 'products_purchased_0',
       'products_purchased_5', 'products_purchased_10']]
y = df['buy']
```

##### Duplicating the features as the model was not performing good enough at first because of bias. To increase the complexity of the model, I'm creating duplicate features by multiplying a constant and raising it to a power of 2.

```python
X=pd.concat([X,X*18,(X*18)**2],axis=1)
```

```python
X
```

##### Splitting the Train and Test dataset and Assigning random state so that the model will produce the same result everytime we reload it

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.50, random_state=18)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression(solver='lbfgs',n_jobs=4,)
logmodel.fit(X_train,y_train)
```

##### Assigning a threshold for the sigmoid function, instead of the defaut threshold I'm assigning 0.35 as the threshold value as it gives better results

```python
threshold=0.35
```

```python
predictions = (logmodel.predict_proba(X_test)[:,1]>=threshold).astype(int)
```

```python
pd.value_counts(predictions)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```

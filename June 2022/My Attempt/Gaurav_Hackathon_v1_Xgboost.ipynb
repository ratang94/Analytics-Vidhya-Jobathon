```python
#Importing Required Stuff
import pandas as pd
import numpy as np
import jinja2
from pycaret.classification import *
import xgboost
```

```python
df_train = pd.read_csv('train.csv')
df_train
```

```python
#Initializing the pycaret model
clf1 = setup(data = df_train, target = 'buy', ignore_features = ['created_at','signup_date'], categorical_features = ['user_activity_var_1','user_activity_var_2',
      'user_activity_var_3','user_activity_var_4','user_activity_var_5','user_activity_var_6','user_activity_var_7','user_activity_var_8','user_activity_var_9',
      'user_activity_var_10','user_activity_var_11','user_activity_var_12','campaign_var_1','campaign_var_2'], numeric_features = ['products_purchased'])
```

```python
#Creating a xgboost model
xgboost = create_model('xgboost')

#Fine tune the created model
tuned_xgboost = tune_model(xgboost)
```

```python
df_test = pd.read_csv('test.csv')
df_test
```

```python
#Predictions with xgboost
predictions = predict_model(tuned_xgboost, data = df_test, raw_score = True)
predictions
```

```python
#Creating Submission file for Hackathon
pred = pd.DataFrame()
pred[['id','buy']] = predictions[['id','Label']]
pred[['id','buy']].to_csv('output.csv',index = False)
```

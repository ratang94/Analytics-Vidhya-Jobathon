```python
#Importing Required Stuff
import pandas as pd
import numpy as np
!pip install -U pycaret scikit-learn
!pip install markupsafe==2.0.1
import jinja2
```

```python
from pycaret.classification import *
import xgboost
```

```python
df_train = pd.read_csv('train.csv')
df_train
```

```python
df_train.isna().sum()
```

```python
df_train['created_at'] = pd.to_datetime(df_train['created_at'], format = "%Y-%m-%d")
df_train['signup_date'] = pd.to_datetime(df_train['signup_date'], format = "%Y-%m-%d")
df_train['created_at'].dtypes
df_train['signup_date'].dtypes

df_train

```

```python
df_train['created_at_years'] = df_train['created_at'].dt.year
df_train['created_at_month'] = df_train['created_at'].dt.month
df_train['created_at_week'] = df_train['created_at'].dt.week
df_train['created_at_day'] = df_train['created_at'].dt.day
df_train['created_at_dayofweek'] = df_train['created_at'].dt.dayofweek

df_train['signup_date_year'] = df_train['signup_date'].dt.year
df_train['signup_date_month'] = df_train['signup_date'].dt.month
df_train['signup_date_week'] = df_train['signup_date'].dt.week
df_train['signup_date_at_day'] = df_train['signup_date'].dt.day
df_train['signup_date_at_dayofweek'] = df_train['signup_date'].dt.dayofweek

df_train
```

```python
s = setup(data = df_train, target = 'buy', session_id=123, ignore_features = ['created_at','signup_date'],numeric_features = ['products_purchased'], numeric_imputation='mean', categorical_imputation='mode', fix_imbalance = True)
```

```python
#Getting the best model
best_model = compare_models(sort = 'F1')
```

```python
print(best_model)
```

```python
#Fine tune the created model
tuned_model = tune_model(best_model)
```

```python
df_test = pd.read_csv('test.csv')
df_test

```

```python
df_test['created_at'] = pd.to_datetime(df_test['created_at'], format = "%Y-%m-%d")
df_test['signup_date'] = pd.to_datetime(df_test['signup_date'], format = "%Y-%m-%d")

df_test['created_at_years'] = df_test['created_at'].dt.year
df_test['created_at_month'] = df_test['created_at'].dt.month
df_test['created_at_week'] = df_test['created_at'].dt.week
df_test['created_at_day'] = df_test['created_at'].dt.day
df_test['created_at_dayofweek'] = df_test['created_at'].dt.dayofweek

df_test['signup_date_year'] = df_test['signup_date'].dt.year
df_test['signup_date_month'] = df_test['signup_date'].dt.month
df_test['signup_date_week'] = df_test['signup_date'].dt.week
df_test['signup_date_at_day'] = df_test['signup_date'].dt.day
df_test['signup_date_at_dayofweek'] = df_test['signup_date'].dt.dayofweek

df_test
```

```python
#Predictions with best_model (Adaboost in this case)
predictions = predict_model(tuned_model, data = df_test, raw_score = True)
predictions
```

```python
#Creating Submission file for Hackathon
pred = pd.DataFrame()
pred[['id','buy']] = predictions[['id','Label']]
pred[['id','buy']].to_csv('output_v4_2.csv',index = False)
```
